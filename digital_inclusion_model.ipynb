# DIGITAL INCLUSION IN HIGH-INCOME COUNTRIES

### Variables:
- **anydigpayment** - made or received a digital payment (1 - yes, 0 - no) - dependent variable
- **economy** - name of the country
- **economycode** - code of the country
- **female** - sex of respondent (1 - female, 2 - male)
- **age** - respondent’s age (years)
- **educ** - respondent education level (1 - primary school or less, 2 - secondary school, 3 - tertiary education or more)
- **inc_q** - within-economy household income quintile (1 to 5)
- **emp_in** - respondent is in workforce (1 - yes, 0 - no)
- **urbanicity** - respondent lives in rural area (1 - yes, 0 - no)
- **dig_account** - has a digitally enabled account (1 - yes, 0 - no)
- **borrowed** - if the respondent personally borrowed money (1 - yes, 0 - no)
- **saved** - if the respondent personally saved or set aside money in the past year (1 - yes, 0 - no)
- **fin2** - has a debit card (1 - yes, 0 - no)
- **fin45** - greatest financial worry (1 - old age, 2 - business costs, 3 - medical costs, 4 - monthly expenses, 5 - education fees, 6 - other, 7 - don’t know, 8 - refused)
- **internet_use** - used the internet within the past three months (1 - yes, 0 - no)

Data source: The Global Findex Database 2025 https://microdata.worldbank.org/index.php/catalog/7860/related-materials
